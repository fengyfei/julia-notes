```julia
using Plots
pyplot()

n = 100
x = range(-10, stop = 10, length = n)
y = range(-10, stop = 10, length = n)
f(x, y) = sin(x) + cos(y)

plot(x, y, f, st = :surface, camera = (-30, 30))
```

```julia
using Plots
pyplot()

n = 100
u = range(0, stop = 2π, length = n)
v = range(0, stop = π, length = n)

x = cos.(u) * sin.(v)'
y = sin.(u) * sin.(v)'
z = ones(n) * cos.(v)'

surf(x, y, z, rstride = 2, cstride = 2)
```
